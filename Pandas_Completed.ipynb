```python
# Dependencies
import pandas as pd
import numpy as np
purchase_data_csv = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(purchase_data_csv)
```

```python
#Total Number of Players
total_players = len(purchase_data["SN"].value_counts())

player_count = pd.DataFrame({"Total Players":[total_players]})
player_count
```

```python
# Number of Unique Items
unique_items_count = len((purchase_data["Item ID"]).unique())
number_of_purchases = (purchase_data["Purchase ID"]).count()
average_price=(purchase_data["Price"]).mean()
total_revenue = (purchase_data["Price"]).sum()
Number_of_Unique_Items = pd.DataFrame({"Unique Items":[unique_items_count],
                           "Purchases": [number_of_purchases], 
                           "Average Price":[average_price], 
                           "Total Revenue": [total_revenue]})
Number_of_Unique_Items
```

```python
#Gender Demographics
Gender_Demographics = purchase_data.groupby("Gender")

total_count_gender = Gender_Demographics.nunique()["SN"]

percentage_of_players = total_count_gender / total_players * 100

gender_percents = pd.DataFrame({"Percentage": percentage_of_players, "Total": total_count_gender})

gender_percents.sort_values(["Total"], ascending = False).style.format({"Percentage":"{:.2f}"})
```

```python
# Purchasing Analysis (Gender):

#Purchase Count
purchase_count = Gender_Demographics["Purchase ID"].count()

# Average Purchase Price
avg_purchase_price = Gender_Demographics["Price"].mean()

# Total Purchase Value 
avg_purchase_total = Gender_Demographics["Price"].sum()

# Average Purchase Total per Person by Gender
avg_purchase_per_person = avg_purchase_total/total_count_gender

 
gender_percents = pd.DataFrame({"Purchase Count": purchase_count, 
                                    "Avg Price": avg_purchase_price,
                                    "Avg Value":avg_purchase_total,
                                    "Avg Total per Person": avg_purchase_per_person})
gender_percents
```

```python
# Age Demographics

#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Average Purchase Total per Person by Age Group
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins, labels=group_names)
purchase_data

age_grouped = purchase_data.groupby("Age Group")

total_count_age = age_grouped["SN"].nunique()

percentage_by_age = (total_count_age/total_players) * 100

age_demographics = pd.DataFrame({"Percentage of Players": percentage_by_age, "Total Count": total_count_age})



age_demographics
```

```python
#Purchase Count
purchase_count_age = age_grouped["Purchase ID"].count()

#Average Purchase Price
avg_purchase_price_age = age_grouped["Price"].mean()

# Calculate total purchase value by age group 
total_purchase_value = age_grouped["Price"].sum()

#Total Purchase Value
avg_purchase_per_person_age = total_purchase_value/total_count_age

age_demographics = pd.DataFrame({"Purchase Count": purchase_count_age,
                                 "Avg Price": avg_purchase_price_age,
                                 "Total Value":total_purchase_value,
                                 "Avg Total per Person": avg_purchase_per_person_age})

age_demographics
```

```python
# Top Spenders
#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

#SN
spender_info = purchase_data.groupby("SN")

# Purchase Count
purchase_count_spender = spender_info["Purchase ID"].count()

# Average Purchase Price
avg_purchase_price_spender = spender_info["Price"].mean()

# Total Purchase Valuel 
purchase_total_spender = spender_info["Price"].sum()


top_spenders = pd.DataFrame({"Purchase Count": purchase_count_spender,
                             "Average Purchase Price": avg_purchase_price_spender,
                             "Total Purchase Value":purchase_total_spender})


big_spendas = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()
big_spendas
```

```python
#Most Popular Items
items = purchase_data[["Item ID", "Item Name", "Price"]]

# Item Name Item ID 
item_info = items.groupby(["Item ID" , "Item Name"])

#Purchase Count 
purchase_count_item = item_info["Price"].count()

# Total Purchase Value 
purchase_value = (item_info["Price"].sum()) 

# Item Price
item_price = purchase_value/purchase_count_item

most_popular_items = pd.DataFrame({"Purchase Count": purchase_count_item, 
                                   "Item Price": item_price,
                                   "Total Value":purchase_value})

popular = most_popular_items.sort_values(["Purchase Count"], ascending=False).head()


popular
```
